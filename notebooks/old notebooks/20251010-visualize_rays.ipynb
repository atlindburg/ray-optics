# Basic Ray

```python
import numpy as np
import matplotlib.pyplot as plt

class Ray:
    def __init__(self, x0=0, y0=0, theta=0):
        """Initialize a 2D ray."""
        self.x0 = x0
        self.y0 = y0
        self.theta = theta  # radians

    def propagate(self, distance):
        """Return a new Ray after propagating forward by 'distance'."""
        new_x = self.x0 + distance
        new_y = self.y0 + distance * np.tan(self.theta)
        return Ray(new_x, new_y, self.theta)

    def path(self, x_end, n_points=100):
        """Return arrays (x, y) of the ray path."""
        x = np.linspace(self.x0, x_end, n_points)
        y = self.y0 + (x - self.x0) * np.tan(self.theta)
        return x, y
```

```python
def visualize_rays(y0_values, theta_values, z_distance):
    """Visualize a set of rays propagating to a distance z."""
    plt.figure(figsize=(8, 4))
    
    for y0 in y0_values:
        for theta in theta_values:
            ray = Ray(y0=y0, theta=np.deg2rad(theta))
            x, y = ray.path(z_distance)
            plt.plot(x, y, 'b', alpha=0.6)
    
    plt.xlabel("x (propagation direction)")
    plt.ylabel("y (height)")
    plt.title("Ray Bundle Propagation")
    plt.grid(True, alpha=0.3)
    plt.axis("equal")
    plt.show()
```

## Visualize a Bunch of Rays

```python
# Define a set of initial ray heights and angles
y0_values = np.linspace(-2, 2, 5)        # mm or arbitrary units
theta_values = np.linspace(-5, 5, 3)     # degrees
z_distance = 20                          # propagation distance

visualize_rays(y0_values, theta_values, z_distance)
```
