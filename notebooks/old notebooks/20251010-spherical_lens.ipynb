```python
import numpy as np
import matplotlib.pyplot as plt

# --- Ray class ---
class Ray:
    def __init__(self, x0=0, y0=0, theta=0, n=1.0, opl=0):
        self.x0 = x0
        self.y0 = y0
        self.theta = theta
        self.n = n
        self.opl = opl

    def propagate(self, x_end):
        """Return arrays for plotting from x0 to x_end."""
        x = np.linspace(self.x0, x_end, 100)
        y = self.y0 + (x - self.x0) * np.tan(self.theta)
        return x, y
```

## Define a surface boundary using the 2D grid

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Surface class ---
class Surface:
    def __init__(self, x_pos, radius):
        self.x_pos = x_pos
        self.radius = radius

    def plot(self):
        theta = np.linspace(0, 2*np.pi, 100)
        x = self.x_pos + self.radius * np.cos(theta)
        y = self.radius * np.sin(theta)
        plt.plot(x, y, 'k')

# Create a surface
surface = Surface(5, 2)

# Plot the surface
plt.figure(figsize=(8, 8))
surface.plot()
plt.axis('equal')
plt.show()
```
